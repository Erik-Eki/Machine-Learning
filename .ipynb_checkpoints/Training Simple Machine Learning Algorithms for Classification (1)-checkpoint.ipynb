# Perceptroni
Perceptroni on kone-oppimismalli, jonka perus-idea on matkia neuronia aivoissa: Se joko ampuu synapseja tai ei.
1. Paino-arvo on alussa joko 0 tai joku pieni numero.
2. Käy iteraatioittain läpi, Antaa ulostulon (*class label = luokan leima*) ja Päivittää paino-arvon.

Jos perceptroni oppii **oikein**, paino-arvo pysyy samana tai nollassa. Jos se oppii **väärin**, paino-arvo menee lähemmäksi positiivista- tai negatiivista kohde-luokkaa (*target class*).
**Perceptroni siis päättää paino-arvon ja sisääntulon perusteella ampuuko neuroni vai ei.**

Perceptronien **yhteneminen** tapahtuu vain jos 2 luokkaa voi erottaa lineaarisesti ja oppimis-aste on tarpeeksi pieni.
Jos luokkia ei voi erotella lineaarisesti, määritellään joko datasetin maksimi iteraatioiden määrä (*epoch*) ja/tai väärinluokittelun toleranssin määrä.



Perusidea siis näyttää tältä:



**x** on dataa, **w** on paino. Näiden avulla lasketaan **netto**, joka syötetään **kynnys**-funktioon. Jos netto-arvo on isompi kuin kynnyksen arvo, käsitellään luokkaa 1, muutoin käsitellään luokka -1.

Negatiivista kynnystä tai painoa kutsutaan puolueellisuus arvoksi (*bias unit*)

#### Esim.
Jos opetus-datan arvo on 0.5 ja tämä luokitellaan väärin -1, tähän liittyvää painoa kasvatettaisiin 1, jotta netto-sisääntulo olisi positiivisempi seuraavalla kerralla kun tätä samaa data-elementtiä käsitellään.

# Alustetaan Perceptroni ja sen metodit

Alustetaan uusi *Perceptroni* objekti joka **oppii datasta** *fit*-metodilla ja **tekee ennustuksia** *predict*-metodilla.

```python
import numpy as np
class Perceptron(object):
    """Perceptron classifier.
     Parameters
     ------------
     eta : float
    Learning rate (between 0.0 and 1.0)
     n_iter : int
    Passes over the training dataset.
     random_state : int
    Random number generator seed for random weight
    initialization.
     Attributes
     -----------
     w_ : 1d-array
     Weights after fitting.
     errors_ : list
    Number of misclassifications (updates) in each epoch.
     ------------
    We append an underscore (_) to attributes that are not created upon
    the initialization of the object, but we do this by calling the object's other methods,
    for example, self.w_.
    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter

        self.random_state = random_state

    def fit(self, X, y):
        """
        Fit training data.

         Parameters
         ----------
         X : {array-like}, shape = [n_examples, n_features]
         Training vectors, where n_examples is the number of
         examples and n_features is the number of features.
         y : array-like, shape = [n_examples]
         Target values.

         Returns
         -------
         self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01,
        size=1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        #Calculate net input
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        #Return class label after unit step
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

Oppimis-aste 𝜂 vaikuttaa lopputulokseen **vain jos painot alustetaan muuksi kuin nolliksi.**

Jos painot alustetaan nolliksi, oppimis-aste vaikuttaa vain paino-vektorin **kokoon**, ei sen **suuntaan**.

#### Tässä esimerkki, missä v2 vektori on puolet v1 vektorin koo'osta ja näiden välinen kulma tulee olemaan aina 0.

```python
#Luodaan vektorit v1 ja v2, missä vektori
v1 = np.array([1, 2, 3])
v2 = 0.5 * v1
#Lasketaan kulma, jonka lopputulos pitäisi olla 0
np.arccos(v1.dot(v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
```

# Harjoitetaan perceptronia Iris datasetillä

```python
import os
import pandas as pd
s = os.path.join('https://archive.ics.uci.edu', 'ml','machine-learning-databases','iris','iris.data')
print('URL:', s)
```

```python
df = pd.read_csv(s,header=None,encoding='utf-8')
df.tail()
```

Seuraavaksi otetaan ensimmäiset 50 Iris-setosa ja 50 Iris-versicolor elementtiä ja muutetaan ne esittämään arvoja -1 ja 1.

```python
import matplotlib.pyplot as plt
import numpy as np

# select setosa and versicolor and assign them to a vector y
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)

# extract sepal length and petal length to a feature matrix
X = df.iloc[0:100, [0, 2]].values

# plot data
plt.scatter(X[:50, 0], X[:50, 1],
color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1],
color='blue', marker='x', label='versicolor')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

Versicolorilla on siis pitemmät terä- ja verholehdet.

Kukat erottuvat nätisti toisistaan, joten lineaarinen erotusraja onnistuu perceptronilla.

```python
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X, y)

#plot errors for each epoch (number of passes over the training dataset) to check whether the algorithm converged and found a decision boundary
plt.plot(range(1, len(ppn.errors_) + 1),
ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
plt.show()
```

Perceptroni lähentyi kuudennella iteraatiolla ja pystyy nyt luokittelemaan harjoitus-kohteita.

# Visualisaatio päätösrajasta

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
    np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                y=X[y == cl, 1],
                alpha=0.8,
                c=colors[idx],
                marker=markers[idx],label=cl,
                edgecolor='black')

plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

# Adaline (ADAptive LInear NEuron)

- Adaline:
    - Vertailee todellisia luokkien leimoja lineaarisen aktivointifunktion jatkuvaan arvioituun ulostuloon.
- Perceptroni:
    - Vertailee todellisia luokkien leimoja arvioituihin.



```python
class AdalineGD(object):
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01,
        size=1 + X.shape[1])

        self.cost_ = []
    
        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = (y - output)
            #Gradient for the weights 1 to m
            self.w_[1:] += self.eta * X.T.dot(errors) #X.T.dot(errors) is a matrix-vector multiplication between our feature matrix and the error vector
            #Gradient for bias unit (zero-weight)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        #Calculate net input
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        #Compute linear activation
        return X

    def predict(self, X):
        #Return class label after unit step
        return np.where(self.activation(self.net_input(X))>= 0.0, 1, -1)
```

Adaline laskee kaltevuuden koko harjoitus datasetille



# Visualisaatio eri oppimisasteiden kustannuksista.

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ada1 = AdalineGD(n_iter=10, eta=0.01).fit(X, y)

ax[0].plot(range(1, len(ada1.cost_) + 1),
np.log10(ada1.cost_), marker='o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log(Sum-squared-error)')
ax[0].set_title('Adaline - Learning rate 0.01')

ada2 = AdalineGD(n_iter=10, eta=0.0001).fit(X, y)
ax[1].plot(range(1, len(ada2.cost_) + 1),
ada2.cost_, marker='o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Sum-squared-error')
ax[1].set_title('Adaline - Learning rate 0.0001')
plt.show()
```

Vasen kuvaaja näyttää kun oppimisaste on liian iso. Kustannukset ei minimoidu vaan errorit vain kasvavat joka epochilla, koska globaali minimi ylitetään.

Oikea kuvaaja taas näyttää kun oppimisaste on liian pieni, kustannukset pienenevät, mutta lähentyminen vaatisi hyvin monta iteraatiota.



# Parannetaan kaltevuuslaskua ominaisuuksien skaalauksella

Standartoinnilla ja normalisoinnilla kaltevuuslaskua voi parantaa.

Tämä siksi, koska optimoija löytää optimaalisimman vastauksen nopeammin, koska painoja päivitetään enemmän.



## Standartointi

```python
X_std = np.copy(X)
X_std[:,0] = (X[:,0] - X[:,0].mean()) / X[:,0].std()
X_std[:,1] = (X[:,1] - X[:,1].mean()) / X[:,1].std()
```

Tällä kertaa käytetään oppimisasteena 0.01

```python
ada_gd = AdalineGD(n_iter=15, eta=0.01)
ada_gd.fit(X_std, y)
plot_decision_regions(X_std, y, classifier=ada_gd)
plt.title('Adaline - Gradient Descent')
plt.xlabel('sepal length [standardized]')
plt.ylabel('petal length [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

plt.plot(range(1, len(ada_gd.cost_) + 1),
ada_gd.cost_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Sum-squared-error')
plt.tight_layout()
plt.show()
```

# Stochastinen kaltevuuslasku (Stochastic Gradient Descent = SGD)

Edellistä tapaa voisi kutsua Erän Kaltevuuslaskuksi (batch gradient descent) joka päivittää painot virheiden summaa mitä harjoituskohteista saadaan.

**SGD** taas päivittää painot asteittain joka harjoituskohteelle.

Adaline algorytmin voi pitää suht-koht ennallaan, tehdään vain pari muutosta *fit*-metodiin, tehdään uusi *partial_fit*-metodi ja lisätään *shuffle*-metodi, joka estää turhan toiston iteraatioissa.

```python
class AdalineSGD(object):
    def __init__(self, eta=0.01, n_iter=10,
                 shuffle=True, random_state=None):
        self.eta = eta
        self.n_iter = n_iter
        self.w_initialized = False
        #Shuffles training data every epoch if True to prevent cycles.
        self.shuffle = shuffle
        #Random number generator seed for random weight initialization.
        self.random_state = random_state

    def fit(self, X, y):
        self._initialize_weights(X.shape[1])
        self.cost_ = []
        for i in range(self.n_iter):
            if self.shuffle:
                X, y = self._shuffle(X, y)
            cost = []
            for xi, target in zip(X, y):
                cost.append(self._update_weights(xi, target))
            avg_cost = sum(cost) / len(y)
            self.cost_.append(avg_cost)
        return self

    def partial_fit(self, X, y):
        '''Fit training data without reinitializing the weights'''
        if not self.w_initialized:
            self._initialize_weights(X.shape[1])
        if y.ravel().shape[0] > 1:
            for xi, target in zip(X, y):
                self._update_weights(xi, target)
        else:
            self._update_weights(X, y)
        return self

    def _shuffle(self, X, y):
        '''Shuffle training data'''
        r = self.rgen.permutation(len(y))
        '''Training Simple Machine Learning Algorithms for Classification'''
        return X[r], y[r]

    def _initialize_weights(self, m):
        '''Initialize weights to small random numbers'''
        self.rgen = np.random.RandomState(self.random_state)
        self.w_ = self.rgen.normal(loc=0.0, scale=0.01,
                                   size=1 + m)
        self.w_initialized = True

    def _update_weights(self, xi, target):
        '''Apply Adaline learning rule to update the weights'''
        output = self.activation(self.net_input(xi))
        error = (target - output)
        self.w_[1:] += self.eta * xi.dot(error)
        self.w_[0] += self.eta * error
        cost = 0.5 * error**2
        return cost

    def net_input(self, X):
        '''Calculate net input'''
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def activation(self, X):
        '''Compute linear activation'''
        return X

    def predict(self, X):
        '''"""Return class label after unit step"""'''
        return np.where(self.activation(self.net_input(X))>= 0.0, 1, -1)
```

```python
ada_sgd = AdalineSGD(n_iter=15, eta=0.01, random_state=1)
ada_sgd.fit(X_std, y)
plot_decision_regions(X_std, y, classifier=ada_sgd)
plt.title('Adaline - Stochastic Gradient Descent')
plt.xlabel('sepal length [standardized]')
plt.ylabel('petal length [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
plt.plot(range(1, len(ada_sgd.cost_) + 1), ada_sgd.cost_,
marker='o')
plt.xlabel('Epochs')
plt.ylabel('Average Cost')
plt.tight_layout()
plt.show()
```

Kustannukset siis laskevat nopeammin SGD:tä käyttäen kuin edellinen erän kaltevuuslasku
